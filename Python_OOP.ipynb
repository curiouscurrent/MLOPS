````python
class Fraction:
  def __init__(self,n,d):
    self.numr = n
    self.denr = d

  def __str__(self):
    # string formatting
    return "{}/{}".format(self.numr,self.denr)

  def __add__(self,other):
    temp_numr = (self.numr * other.denr) + (other.numr * self.denr)
    temp_denr = (self.denr * other.denr)

    return "{}/{}".format(temp_numr,temp_denr)
````

##### list

````python
def change(L):
  # before editing
  print(id(L))
  L.append(5)
  # after editing
  print(id(L))

````

````python
L1 = [1,2,3,4]
# before making the function call
print(id(L1))
change(L1)
# after making the function call
print(id(L1))
# is the list elements changed
print(L1)
````

if you don't want to make changes in the original we need to do clonning.


````python
def change(L):
  # before editing
  print(id(L))
  L.append(5)
  # after editing
  print(id(L))

L1 = [1,2,3,4]
# before making the function call
print(id(L1))
# clone of L1 is passed, so it will be a new address
change(L1[:])
# after making the function call
print(id(L1))
# is the list elements changed
print(L1)

````

When you do pass by reference,
1. if you pass mutable datatypes/objects, changes will be made in original
2. if you pass immutable dataypes/objects, changes will not be made in original.

### static method -> to access this method, we dont need objects

-> used when dealing with static/class variables

-> self is not passed (self is an object) and object is not required to access the class variable

-> we just need className

-> denoted by keyword @staticmethod

````python
class Atm:

  # static/class variables
  __counter = 1
  def __init__(self): # __init__ is a constructor
    # these variables will be declared in memory
    # instance variables
    self.__pin = ""
    self.__balance = 0
    self.sno = Atm.__counter
    Atm.__counter +=1
    # print(id(self))

    # calling this method.
    # self.menu()
  @staticmethod
  def get_counter():
    return Atm.__counter

  @staticmethod
  def set_counter(new):
    if type(new) == int:
      Atm.__counter = new
    else:
      print("Not allowed")

  def menu(self):
    user_input = input("""
                     Hello, how would you like to proceed ?
                     1. Enter 1 to create pin
                     2. Enter 2 to deposit
                     3. Enter 3 to withdraw
                     4. Enter 4 to check balance
                     5. Enter 5 to exit
    """)

    if user_input == "1":
      self.create_pin()
    elif user_input == "2":
      self.deposit()
    elif user_input == "3":
      self.withdraw()
    elif user_input == "4":
      self.check_balance()
    else:
      print("Exit")

  def create_pin(self):
    self.__pin = input("Enter your pin")
    print("Pin set successfully")

  def deposit(self):
    temp = input("Enter your pin")
    if temp == self.__pin:
      amt = int(input("Enter the amount to deposit"))
      self.__balance = self.__balance + amt
      print("Deposit successful")
    else:
      print("Invalid pin")

  def withdraw(self):
    temp = input("Enter your pin")
    if temp == self.__pin:
      amt = int(input("Enter the amount to withdraw"))
      if amt <= self.__balance:
        self.__balance = self.__balance - amt
        print("Withdraw successful")
      else:
        print("Insufficient funds")

    else:
      print("Invalid pin")

  def check_balance(self):
    temp = input("Enter your pin")
    if temp == self.__pin:
      print("Your balance is : ",self.__balance)
    else:
      print("Invalid pin")
````

# Method overloading

1. technically doesnt exist in python but can be achieved by setting default arguments.

2. if you have the same method written one after the other , the second method will override the first method.

(different inputs -> different behaviours -> one method)

(ex : calculating area -> method "area", inputs for circle -> radius, inputs for rectangle -> length,breadth)

````python
class Geometry:

  def area(self,radius):
    return 3.14 * radius * radius

  def area(self,l,b):
    return l * b
````

````python
gm = Geometry()
gm.area(4)
````

workaround

````python
class Geometry:

  def area(self,l,b=0):
    if b==0:
      return 3.14 * l * l
    else:
      return l * b
````

````python
gm = Geometry()
print(gm.area(4))
print(gm.area(4,5))
````
