```python
import numpy as np
import pandas as pd
deliveries = pd.read_csv('deliveries.csv')
```

```python
deliveries.head()
```

```python
# find the isin() function
# find the most destructive death over batsman in the history of IPL
# min batsman 200 balls in over 16-20 (called the death over)
deliveries.info()
```

```python
# to get overs 16-20
mask = deliveries['over']>15
death_over = deliveries[mask]
```

```python
death_over
```

```python
all_batsman = death_over.groupby('batsman')['batsman_runs'].count()
# we need batsman who has balls > 200
mask_batsman = all_batsman>200
mask_batsman
```

```python
all_batsman
```

```python
# we have all batsman who made > 200 balls in death over (overs 16 to 20)
batsman_list = all_batsman[mask_batsman].index.to_list()
```

```python
# now to calculate the strike rate we need to get the total number of runs and total
# number of balls played by each batsman in overs 16-20
# instead of having 43 masks, we will pass a list to the isin() function
mask = death_over['batsman'].isin(batsman_list)
mask
```

```python
final_df = death_over[mask]
```

```python
runs = final_df.groupby('batsman')['batsman_runs'].sum()
```

```python
runs
```

```python
balls = final_df.groupby('batsman')['batsman_runs'].count()
balls
```

```python
# strike rate
strike_rate = (runs/balls) * 100
strike_rate.sort_values(ascending=False).head(3)
```

```python
# we can see that there are missing values in Age,Cabin and Embarked columns
# for dealing with this, we have dropna() and fillna()
# dropna() = drops the missing values
# fillna() = fills the missing values
# important param is how
# when how = 'any', it will iterate row by row, and in every row it will check if
# there is a column with a missing value. If yes, the entire row will be deleted.
# when how = 'all', it will only drop a row when values in all the columns is missing
# by default axis=0, so drop happens in row.
# if we want the drop for columns, set axis=1
titanic.dropna(how='any').shape
```

```python
titanic.dropna(how='all').shape
```

```python
# now let us set axis=1, drop all columns where there is atleast one missing value in
# that column
# so age,cabin,embarked to be dropped -> 3 columns dropped
titanic.dropna(axis=1,how='any').shape
```

```python
titanic.dropna(axis=1,how='all').shape
```

```python
# subset param in dropna()
# pass a list of one/more columns to the subset param
# lets say we need to drop rows which have value for Cabin missing
# since we have only 204 values for cabin, we will be having 204 rows
cabin = titanic.dropna(subset=['Cabin'])
```

```python
embarked = cabin.dropna(subset=['Embarked'])
```

```python
age = embarked.dropna(subset=['Age'])
age.shape
```

```python
# here an OR operation is performed
# drop the rows where either the value for cabin is missing or
# the value for Embarked is missing
titanic.dropna(subset=['Cabin','Embarked'],how='any',axis=0).shape
```

```python
# cabin (204 values)
# embarked (889 values)
# age (714 values)
titanic.dropna(subset=['Cabin','Embarked','Age'],how='any',axis=0).shape
```

```python
# fillna()
# param = value to be specified, and a method to be specified
# since all columns are not of the same datatype we can't directly do this
titanic.fillna(0)
```

```python
titanic['Cabin']
```

```python
# perform fillna() on a specific column
titanic['Cabin'].fillna("Not specified")
```

```python
# methods are bfill, ffill
# ffill : propagate last valid value forward
# bfill : propagate next valid value backward
titanic['Cabin'].fillna(method='bfill')
```

```python
titanic['Cabin'].fillna(method='ffill')
```
